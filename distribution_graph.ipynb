```python
%matplotlib notebook
import numpy as np
import time
import matplotlib 
import matplotlib.pyplot as plt
from array import array

start_time = time.time()


inputfile = "/big6_disk/fangren106/mock_community/v2_zymo/even/alignment/Big/uniq_map.paf"
tref={}

with open(inputfile,'r') as f:
    for line in f:
        llist = line.split("\t")
        rlen = int(llist[1])
        name = llist[5].split("-")[1]
        name = name.split("_")[0]
        if name not in tref:
            tref[name]=[rlen]
        else:
            tref[name].append(rlen)
linec="black"
for name,arr in tref.items():
    arr.sort()
    scales=[]
    counts=[]
    scale=100
    scale_count=0
    for i in arr:
        if i > 20000: continue
        if i <= scale:
            scale_count += 1
        else:
            scales.append(scale)
            counts.append(scale_count)
            scale += 500
            scale_count = 1
    
    if name == "Bacillus":
        plt.plot(scales,counts,color="blue",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Lactobacillus":
        plt.plot(scales,counts,color="#FF8D33",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Listeria":
        plt.plot(scales,counts,color="#FFC300",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Enterococcus":
        plt.plot(scales,counts,color="#FFDD33",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Staphylococcus":
        plt.plot(scales,counts,color="#C28A13",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Escherichia":
        plt.plot(scales,counts,color="#0BD320",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Salmonella":
        plt.plot(scales,counts,color="#64EF37",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Pseudomonas":
        plt.plot(scales,counts,color="#59F5A3",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Saccharomyces":
        plt.plot(scales,counts,color="#D759F5",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    elif name == "Cryptococcus":
        plt.plot(scales,counts,color="#E337BC",label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    else:
        plt.plot(scales,counts,label=name)
        plt.xticks(range(0,21000,1000),rotation=90)
    plt.legend(loc='upper right')    
    plt.show()
            
```
